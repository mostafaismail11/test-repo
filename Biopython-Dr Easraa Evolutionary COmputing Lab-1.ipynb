```python
# you need to collect all the species which are present in each record 
# how can we do it ?
#we can do it using loop


species=[]

for record in SeqIO.parse("/home/mostafa/Downloads/ls_orchid.gbk",
                                                 'genbank'):
    species.append(record.annotations['organism'])
    
print(species)
    
```
